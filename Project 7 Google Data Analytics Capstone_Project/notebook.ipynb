# Clean Data and Prepare them for analysis

```python
bikes_annual.info()
```

#### Convert the data type of started_at and ended_at to datetime

```python
bikes_annual['started_at'] = pd.to_datetime(bikes_annual['started_at'], format='%Y-%m-%d %H:%M:%S')
```

```python
bikes_annual['ended_at'] = pd.to_datetime(bikes_annual['ended_at'], format='%Y-%m-%d %H:%M:%S')
```

```python
bikes_annual.info()
```

#### I will extract the time from both the started_at and ended_at columns, to culculate the lenght of the ride in minutes

```python
bikes_annual['started_time'] = bikes_annual['started_at'].dt.time
```

```python
bikes_annual['ended_time'] = bikes_annual['ended_at'].dt.time
```

```python
bikes_annual['started_time'] = bikes_annual['started_time'].astype('str')
```

```python
bikes_annual['started_time_min'] = bikes_annual['started_time'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]) + int(x[2]) / 60)
```

```python
bikes_annual['ended_time'] = bikes_annual['ended_time'].astype('str')
bikes_annual['ended_time_min'] = bikes_annual['ended_time'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]) + int(x[2]) / 60)
```

```python
bikes_annual.head()
```

```python
bikes_annual.info()
```

```python
bikes_annual['ride_length'] = bikes_annual['ended_time_min'] - bikes_annual['started_time_min']
```

```python
bikes_annual.head()
```

# EDA

```python
bike_ride = pd.read_csv(r'csv_files/cleaned_data.csv')
```

```python
bike_ride.head()
```

```python
# remove the first column (Unnamned: 0)
bike_ride.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
bike_ride.head()
```

```python
bike_ride.describe()
```

```python
bike_ride_av = bike_ride.groupby('member_casual').agg({'ride_length_min' : 'mean'})
```

#### Member clients are more than casual clients by 546282

```python
bike_ride[bike_ride['member_casual'] == 'member']
# the member clients are 2615035
```

```python
bike_ride[bike_ride['member_casual'] == 'casual']
# the casual members are 2068753
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='member_casual', data=bike_ride)

plt.title('Casual vs Members')
```

#### The average length of the casual's rides is bigger than the average length of the member's rides according to the following graph

```python
bike_ride_av.plot(kind='bar')
plt.title('Average ride length')
```

#### We observe a huge difference in the usage of classic bikes. Members use this type more than cassuals.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='rideable_type', data=bike_ride, hue='member_casual')

plt.title('Types of bikes for each kind of client')
```

#### Saturday is the most preferred day for cycling, following by Sunday and Friday from all the clients

```python
plt.figure(figsize=(8,5))

sns.countplot(x='weekday', data=bike_ride)

plt.title('The most preferred days for cycling')
```

#### Casual clients prefer riding a bike at the weekends and specifically on Saturdays. While member clients prefer riding a bike all week except Sunday

```python
plt.figure(figsize=(8,5))

sns.countplot(x='weekday', data=bike_ride, hue='member_casual')

plt.title('The most preferred days for cycling of each type of client')
```

#### The member clients ride the bikes for less time than the casual clients on every day of the week

```python
plt.figure(figsize=(8,5))

sns.barplot(x='weekday', y='ride_length_min', data=bike_ride, hue='member_casual')

plt.title('Length of rides per day')
```

# <center>Conclusions</center>
<ul>
    <li>Member clients are more than casual clients</li>
    <li>The average length of the casual's rides is bigger than the average length of the member's rides</li>
    <li>Member clients prefer classic bikes more than the other types</li>
